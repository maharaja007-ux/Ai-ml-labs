```python
n = int(input("Enter number of nodes: "))

graph = {}

for i in range(n):
    node = input("Enter node: ")
    neighbors = input("Enter neighbors separated by space: ").split()
    graph[node] = neighbors

root_node = input("Enter The Root Node: ")
goal_node = input("Enter The Goal Node: ")

visited = set()

def dfs(graph, node, goal_node):
    if node in visited:
        return False

    print(node, end=" ")
    visited.add(node)

    if node == goal_node:
        print(f"\nGoal node found - {goal_node}")
        

    for neighbour in graph[node]:
        if dfs(graph, neighbour, goal_node):
            return True

    return False


print("DFS Traversal Order:-")
dfs(graph, root_node, goal_node)
```

```python
n = int(input("Enter number of nodes: "))
graph = {}

for i in range(n):
    node = input("Enter node: ")
    neighbors = input("Enter neighbors separated by space: ").split()
    graph[node] = neighbors

Root_node = input("Enter The Root Node: ")
Goal_node = input("Enter The Goal Node: ")

print(graph)

def bfs(graph, start, goal):
    visited = []
    queue = []
    parent = {}   

    visited.append(start)
    queue.append(start)
    parent[start] = None

    while queue:
        m = queue.pop(0)
        print(m, end=" ")

        if m == goal:
            print(f"\nGoal Node found - {goal}")

           
            path = []
            current = goal
            while current is not None:
                path.append(current)
                current = parent[current]

            path.reverse()
            print("Shortest Path:", " -> ".join(path))
            return

        for neighbour in graph[m]:
            if neighbour not in visited:
                visited.append(neighbour)
                queue.append(neighbour)
                parent[neighbour] = m

    print("\nGoal node not found")


print("BFS Traversal Order:")
bfs(graph, Root_node, Goal_node)
```

```python
import networkx as nx

graph = nx.Graph()

while True:
    choice = int(input("Insert edge? Press 1 for yes, 2 for no: "))
    
    if choice == 1:
        node1 = input("Enter node1: ").strip().upper()
        node2 = input("Enter node2: ").strip().upper()
        weight = int(input("Enter weight between nodes: "))
        
        graph.add_edge(node1, node2, weight=weight)
    else:
        break

print("\nGraph adjacency list:")
for node, neighbors in graph.adj.items():
    print(node, neighbors)

start = input("\nEnter start node: ").strip().upper()
goal = input("Enter goal node: ").strip().upper()

heuristic_values = {}
heuristic_values[goal] = 0

for node in graph.nodes():
    if node != goal:
        h = int(input(f"Enter heuristic value from {node} to {goal}: "))
        heuristic_values[node] = h

def heuristic(u, v):
    return heuristic_values[u]

path = nx.astar_path(graph, start, goal, heuristic=heuristic, weight='weight')
cost = nx.astar_path_length(graph, start, goal, heuristic=heuristic, weight='weight')

print("\nShortest Path:")
print(" -> ".join(path))

print("Total Cost:", cost)
```
