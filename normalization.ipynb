```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
```

```python
info={
    "Types_Of_Residence_(BHK)":[1,2,3,4,1,2,3,2,1,2,2,1,1,4,2],
    "Area_Of_Residence(Sqr-feet)" :[3000,6000,9000,12000,3000,6000,9000,6000,3000,6000,6000,3000,3000,12000,6000],
    "Locality_Of_Residence":["RaniPokhari","doiwala","rampur","tehri","dehradun","rishikesh","chamoli","almora","sharanput","haridwari","mobewala","poantasahib","chamba","haridwar","rampur"],
    "Price_Of_Residence":[150000,300000,450000,600000,150000,300000,450000,300000,150000,300000,300000,150000,150000,400000,300000],
    "No_Of_floor":[2,3,5,8,2,5,6,4,5,2,1,2,3,5,2]
}
df1=pd.DataFrame(info)
df1
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df1[['scaled_area','scaled_price']] = scaler.fit_transform(
    df1[['Area_Of_Residence(Sqr-feet)', 'Price_Of_Residence']]
)

print(df1.head())
```

```python
print(df1[["Area_Of_Residence(Sqr-feet)",'Price_Of_Residence']])
```

```python
plt.figure()
plt.hist(df1["scaled_area"])
plt.title("Normalized BHK Distribution")
plt.show()
```

```python
plt.figure()
plt.scatter(df1["scaled_area"], df1["scaled_price"])
plt.xlabel("Scaled BHK")
plt.ylabel("Scaled Price")
plt.title("Scaled BHK vs Scaled Price")
plt.show()
```

```python
area_min = df1["Area_Of_Residence(Sqr-feet)"].min()
area_max = df1["Area_Of_Residence(Sqr-feet)"].max()

df1["Area_Normalized"] = (df1["Area_Of_Residence(Sqr-feet)"] - area_min) / (area_max - area_min)

cost_min = df1["Price_Of_Residence"].min()
cost_max = df1["Price_Of_Residence"].max()

df1["Purchasing_Cost_Normalized"] = (df1["Price_Of_Residence"] - cost_min) / (cost_max - cost_min)
df1
```

```python
sns.scatterplot(x=df1["Area_Normalized"], y=df1["Purchasing_Cost_Normalized"])
plt.title("After Min-Max Normalization")
plt.xlabel("Normalized Area")
plt.ylabel("Normalized Cost")
plt.show()
```

```python
area_mean = df1['Area_Of_Residence(Sqr-feet)'].mean()
area_std = df1['Area_Of_Residence(Sqr-feet)'].std()

price_mean = df1['Price_Of_Residence'].mean()
price_std = df1['Price_Of_Residence'].std()

# Apply standardization
df1['std_area'] = (df1['Area_Of_Residence(Sqr-feet)'] - area_mean) / area_std
df1['std_price'] = (df1['Price_Of_Residence'] - price_mean) / price_std
```

```python
plt.figure()
plt.scatter(df1['std_area'], df1['std_price'])
plt.xlabel("Standardized Area")
plt.ylabel("Standardized Price")
plt.title("Manual Standardization (Area vs Price)")
plt.show()
```
