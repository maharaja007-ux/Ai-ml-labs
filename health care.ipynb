```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load CSV file
df = pd.read_csv("C:\\Users\\ritik\\OneDrive\\Desktop\\insurance1.csv")

plt.figure()
plt.scatter(df["age"], df["charges"])
plt.xlabel("Age")
plt.ylabel("Charges")
plt.title("Age vs Insurance Charges")
plt.show()


```

```python
df.info()
df.describe()
df
```

```python
smoker_counts = df["smoker"].value_counts()

plt.figure()
plt.pie(smoker_counts.values, labels=smoker_counts.index, autopct="%1.1f%%")
plt.title("Smoker vs Non-Smoker Distribution")
plt.show()
```

```python
plt.figure()
plt.scatter(df["bmi"], df["charges"])
plt.xlabel("BMI")
plt.ylabel("Charges")
plt.title("BMI vs Insurance Charges")
plt.show()
```

```python
avg_smoker = df.groupby("smoker")["charges"].mean()
plt.figure()
plt.bar(avg_smoker.index, avg_smoker.values)
plt.xlabel("Smoker")
plt.ylabel("Average Charges")
plt.title("Average Charges: Smoker vs Non-Smoker")
plt.show()
```

```python
avg_children = df.groupby("children")["charges"].mean()
plt.figure()
plt.plot(avg_children.index, avg_children.values)
plt.xlabel("Number of Children")
plt.ylabel("Average Charges")
plt.title("Average Charges vs Number of Children")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=[np.number])
corr = numeric_df.corr()

plt.figure()
plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Heatmap of Insurance Dataset")
plt.show()
```

```python
smoker_yes = df[df["smoker"] == "yes"]["charges"]
smoker_no = df[df["smoker"] == "no"]["charges"]


plt.figure()
plt.boxplot([smoker_no, smoker_yes], labels=["Non-Smoker", "Smoker"])
plt.xlabel("Smoker Category")
plt.ylabel("Insurance Charges")
plt.title("Boxplot of Charges by Smoker Status")
plt.show()
```
