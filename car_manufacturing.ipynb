```python
df2=df.copy()
```

```python
df2.isnull().sum()
```

```python
df2
```

```python
age_milage=df2["milage"].mean()
df2["milage"]=df2["milage"].fillna(age_milage)
df2["colour"]=df2["colour"].fillna("white")
df2["car_model"]=df2["car_model"].fillna("base")
type_=df2["Type"].mode()[0]
df2["Type"]=df2["Type"].fillna(type_)
df2["mfg_year"]=df2["mfg_year"].fillna("2026").astype("int64")
engine_capacity=df2["engine_capacity"].mean()
df2["engine_capacity"]=df2["engine_capacity"].fillna(engine_capacity)
df2["price(LAKHS)"]=df2["price(LAKHS)"].fillna(20)
df2
```

```python
df2
```

```python
plt.pie(df2["Type"].value_counts(),labels=df2["Type"].value_counts().index,autopct="%1.1f%%")
plt.title("Distrbution of cars by fuel type")
plt.show()
```

```python
avg_price = df2.groupby("Type")["price(LAKHS)"].mean()

plt.figure()
plt.bar(avg_price.index, avg_price.values)
plt.xlabel("Car Type")
plt.ylabel("Average Price (Lakhs)")
plt.title("Average Price by Fuel Type")
plt.show()
```

```python

df_sorted = df2.sort_values("mfg_year")

plt.figure()
plt.plot(df_sorted["mfg_year"], df_sorted["price(LAKHS)"], marker='o')
plt.xlabel("Manufacturing Year")
plt.ylabel("Price (Lakhs)")
plt.title("Price Trend Over Years")
plt.show()
```

```python
plt.figure()
plt.scatter(df2["milage"], df2["price(LAKHS)"])
plt.xlabel("Mileage")
plt.ylabel("Price (Lakhs)")
plt.title("Mileage vs Price")
plt.show()
```

```python
avg_milage = df.groupby("car_mfg")["milage"].mean()

plt.figure(figsize=(10,5))   # increased width
plt.bar(avg_milage.index, avg_milage.values)
plt.xlabel("Car Manufacturer")
plt.ylabel("Average Mileage")
plt.title("Average Mileage by Manufacturer")
plt.xticks(rotation=60, ha="right")  # better rotation
plt.tight_layout()
plt.show()
```
